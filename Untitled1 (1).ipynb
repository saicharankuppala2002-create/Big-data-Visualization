```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Teen_Mental_Health_Dataset.csv.xlsx')

print(df.head())

plt.figure(figsize=(6,4))
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
df['platform_usage'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Social Media Platform Usage")
plt.ylabel("")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df['sleep_hours'], bins=10)
plt.title("Sleep Hours Distribution")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['daily_social_media_hours'],
            df['academic_performance'])

plt.title("Social Media vs Academic Performance")
plt.xlabel("Daily Social Media Hours")
plt.ylabel("Academic Performance")
plt.show()


plt.figure(figsize=(6,4))
df['physical_activity'].plot(kind='line')

plt.title("Physical Activity Trend")
plt.xlabel("Index")
plt.ylabel("Physical Activity Hours")
plt.show()

plt.figure(figsize=(6,4))
plt.boxplot(df['screen_time_before_sleep'])

plt.title("Screen Time Before Sleep")
plt.ylabel("Hours")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=8)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(8,6))
plt.imshow(numeric_df.corr(), cmap='coolwarm')

plt.colorbar()
plt.xticks(range(len(numeric_df.columns)),
           numeric_df.columns,
           rotation=90)

plt.yticks(range(len(numeric_df.columns)),
           numeric_df.columns)

plt.title("Correlation Heatmap")
plt.show()
```
